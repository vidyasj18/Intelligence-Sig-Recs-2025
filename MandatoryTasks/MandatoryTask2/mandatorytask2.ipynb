```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# loading the data from dataset
```

```python
train_df = pd.read_csv("train_data.csv")
test_df = pd.read_csv("test_data.csv")
```

```python
train_df.shape
```

```python
test_df.shape
```

```python
# here, w and x acts as input features in train and test datasets and y is output feature in train dataset.
```

```python
x_train = train_df[['w','x']]
y_train = train_df['y']

x_test = test_df[['w','x']]
```

```python
# training simple linear regression model on the training dataset
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
# predicting outputs of test set
```

```python
y_test_pred = regressor.predict(x_test)
```

```python
# visualising training set results in 2D
```

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('y vs w and x (Training set)')
plt.xlabel('w and x')
plt.ylabel('y')
plt.show()
```

```python
# visulaising training set results in 3D
```
